```python
#Importamos módulo
import wooldridge as woo
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import datetime as dt
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt
# graficos incrustados
get_ipython().run_line_magic('matplotlib', 'inline')
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

```

```python
# Leer el archivo de Excel
base_datos = pd.read_excel('Base Datos.xlsx')

# Mostrar las primeras filas del DataFrame
print(base_datos.head())
```

```python
#Estadistica descriptiva
summary = base_datos.select_dtypes(include=[np.number]).describe()
print(summary)

#Establecer formato adecuado a la fecha
T= len(base_datos)
base_datos.index= pd.date_range(start='2001', periods=T, freq='A')
base_datos.head()
```

```python
#Grafico de mi variable Real Goverment
plt.plot('PrevDN', data=base_datos, color='black', linestyle='-')
plt.title ("Prevalencia de desnutrición %")
plt.ylabel('% Desnutrición')
plt.xlabel('Fecha')
plt.show()

#Grafico de la FAS
plot_acf(base_datos.index)
plt.title ('Autocorrelación Simple')
plt.ylabel('Rezagos k')
plt.xlabel('Coeficiente de correlación')
plt.show()

plot_pacf(base_datos.index)
plt.title ('Autocorrelación parcial')
plt.ylabel('Rezagos k')
plt.xlabel('Coeficiente de correlación')
plt.show()
```
